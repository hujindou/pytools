```python
file1 = 'c:/dlib/test/hjd.xml'
#fs_read = cv2.FileStorage(file1, cv2.FILE_STORAGE_READ)
tree = ET.parse(file1)
root = tree.getroot()
vecs = []
for type_tag in root.findall('matdata/data'):
    vecs.append(type_tag.text.split())
    #value = type_tag.get('foobar')
    #type_tag.text.split()
    #print(np.array(type_tag.text.split()))
    pass
df1 = pd.DataFrame(vecs).astype('float32')
df1["YLabel"] = 0
```

```python
file2 = 'c:/dlib/test/jiang.xml'
tree = ET.parse(file2)
root = tree.getroot()
vecs = []
for type_tag in root.findall('matdata/data'):
    vecs.append(type_tag.text.split())
    #value = type_tag.get('foobar')
    #type_tag.text.split()
    #print(np.array(type_tag.text.split()))
    pass
df2 = pd.DataFrame(vecs).astype('float32')
df2["YLabel"] = 1
```

```python
df = pd.concat([df1, df2], ignore_index=True)
```

```python
nparr = df.values
```

```python
resmatrix = nparr[:,np.newaxis,:] - nparr[np.newaxis,:,:]
```

```python
finalmatrix = np.linalg.norm(resmatrix,axis=-1)
```

```python
tmppd = pd.DataFrame(finalmatrix)
```

```python
analyzepd = tmppd.mask(np.triu(np.ones(tmppd.shape, dtype=np.bool_)))
```

```python
analyzepd
```

```python
distvec = []
for tmpi in range(analyzepd.shape[0]):
    for tmpj in range(analyzepd.shape[1]):
        if not math.isnan(analyzepd.iloc[tmpi,tmpj]):
            distvec.append(analyzepd.iloc[tmpi,tmpj])
            pass
        pass
    pass
```

```python
finalseries = pd.Series(distvec)
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(df.iloc[:,0:128], df.iloc[:,128],
                                                random_state=42)
```

```python
from sklearn.svm import SVC # "Support vector classifier"
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()                       # 2. instantiate model
model.fit(Xtrain, ytrain)                  # 3. fit model to data

#model = SVC(C=1.0, kernel="linear", probability=True)
#model.fit(Xtrain, ytrain)
```

```python
y_model = model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

```python
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(ytest, y_model)

sns.heatmap(mat, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value');
```
